### Loading the Cleaned Dataset

```python
import pandas as pd
dataset=pd.read_csv('/content/Fuel_type_cleaned.csv')
print(dataset)
```

### Checking Values

```python
print(dataset["Transmission"].isnull().sum())
print(dataset["Transmission"].dtype)
print(dataset["Transmission"].unique())
print(dataset["Transmission"].value_counts())
```

### Cleaning and Standardizing Data

```python
dataset["Transmission"] = dataset["Transmission"].str.strip()
dataset["Transmission"] = dataset["Transmission"].str.title()
print(dataset["Transmission"].unique())
print(dataset["Transmission"].value_counts())
```

### Verifying the Cleaned Transmission Column

```python
print(dataset["Transmission"].isnull().sum())
print(dataset["Transmission"].dtype)
print(dataset["Transmission"].unique())
print(dataset["Transmission"].duplicated().sum())
print(len(dataset["Transmission"]))
```

### Saving the Updated Dataset

```python
from google.colab import files
dataset.to_csv('Transmission_cleaned.csv',index=False)
files.download('/content/Transmission_cleaned.csv')
```
