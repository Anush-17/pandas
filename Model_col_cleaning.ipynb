### Loading the Cleaned Dataset

```python
import pandas as pd
dataset=pd.read_csv('/content/Price_cleaned.csv')
print(dataset)
```

### Checking the Missing Values

```python
print(dataset["Model"].isnull().sum())
print(dataset["Model"].dtype)
print(dataset["Model"].unique())
print(dataset["Model"].value_counts())
```

### Cleaning and Standardizing Model Names

```python
dataset["Model"] = dataset["Model"].str.strip()
dataset["Model"] = dataset["Model"].str.title()
print(dataset["Model"].unique())
print(dataset["Model"].value_counts())
```

### Correcting Inconsistent Model Names

```python
dataset["Model"] = dataset["Model"].replace({
    "Gla"  : "GLA",
    "Glc"  : "GLC",
    "Cr-V" : "CR-V",
    "Rav4" : "RAV4"
})
print(dataset["Model"].unique())
```

### Reviewing the Cleaned Model Data

```python
print(dataset["Model"].isnull().sum())
print(dataset["Model"].dtype)
print(dataset["Model"].unique())
print(dataset.duplicated().sum())
print(len(dataset))
```

### Saving the Final Dataset

```python
from google.colab import files
dataset.to_csv('Model_cleaned.csv',index=False)
files.download('/content/Model_cleaned.csv')
```
