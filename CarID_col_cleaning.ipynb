### Loading the Cleaned Dataset

```python
import pandas as pd
dataset=pd.read_csv('/content/Model_cleaned.csv')
print(dataset)
```

### Checking Car ID Column Details

```python
print(dataset["Car ID"].isnull().sum())
print(dataset["Car ID"].dtype)
print(dataset["Car ID"].describe())
print(dataset["Car ID"].head(10))
```

### Converting Car ID to Integer Format

```python
dataset["Car ID"] = dataset["Car ID"].astype(int)
print(dataset["Car ID"].dtype)
print(dataset["Car ID"].head(10))
```

### Reviewing the Dataset After Conversion

```python
print(dataset["Car ID"].isnull().sum())
print(dataset["Car ID"].dtype)
print(dataset["Car ID"].min())
print(dataset["Car ID"].max())
print(dataset.duplicated().sum())
print(len(dataset))
```

### Saving the Final Dataset

```python
from google.colab import files
dataset.to_csv('Car_price_fully_cleaned.csv',index=False)
files.download('/content/Car_price_fully_cleaned.csv')
```
