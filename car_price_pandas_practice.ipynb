### Load a Dataset

```python
import pandas as pd
dataset=pd.read_csv("/content/car_price_prediction_with_missing.csv")
print(dataset.to_string())
```

### Understanding the Dataset

```python
print(dataset.shape)
print(dataset.info())
print(dataset.describe().to_string())
```

### Display unique values in categorical columns

```python
print(dataset['Brand'].unique())
print(dataset['Fuel Type'].unique())
print(dataset['Transmission'].unique())
print(dataset['Condition'].unique())
print(dataset['Model'].unique())
print("\n Count the values of Model \n")
print(dataset['Model'].value_counts()) #count the values of a column
```

### Display Selected Data

```python
print(dataset[['Brand','Fuel Type','Transmission','Condition','Model']])
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

### Text Transformation

```python
dataset['Brand']=dataset['Brand'].str.lower()
print(dataset['Brand'].head(5))
print()
dataset['Brand']=dataset['Brand'].str.upper()
print(dataset['Brand'].head(5))
print()
dataset['Brand']=dataset['Brand'].str.title()
print(dataset['Brand'].head(5))
print()
```

### Filtering & Selecting Data

```python
print('First five rows of Tesla Brands \n')
tesla_cars=dataset[dataset['Brand'] =='Tesla']
print(tesla_cars.head(5).to_string())
print('\nFirst five rows of Manual Type Cars \n')
Manual_cars=dataset[dataset['Transmission'] == 'Manual']
print(Manual_cars.head(5).to_string())
```

### Checking Missing and Available Values

```python
print(dataset['Brand'].isnull().sum())
print(dataset['Brand'].notnull().sum())
```

### Handling Missing Values

```python

null_val=dataset.fillna('null').to_string()
print(null_val)
print(dataset.dropna(axis=0).tail(10).to_string())
print(dataset.dropna(axis=1).tail(10).to_string())
print("\n count of Nulls \n")
print(dataset['Model'].value_counts(dropna=False).to_string())
print("\n Drop Null and count of NotNull \n")
print(dataset['Model'].value_counts(dropna=True).to_string())
```
