## Import Dataset

```python
import pandas as pd
dataset=pd.read_csv('/content/Brand_col_cleaned.csv')
print(dataset)
```

## Check Missing Values

```python
print(dataset['Year'].isnull().sum())
```

## Convert Year Column Data Type to Integer

```python
print(dataset['Year'].dtype)
dataset['Year']=dataset['Year'].astype(int)
print(dataset['Year'].dtype)
```

## Display Unique Values

```python
print(sorted(dataset['Year'].unique()))
```

## Find Min and Max Year Values

```python
print("Minimum year:",dataset['Year'].min())
print("Maximum year:",dataset['Year'].max())
```

## Data Quality

```python
print(dataset['Year'].isnull().sum())
print(dataset['Year'].duplicated().sum())
print(dataset['Year'].dtype)
print(len(dataset['Year']))
print(dataset['Year'])
```
