## Brand Column Data Cleaning

### Load a Dataset

```python
import pandas as pd
dataset=pd.read_csv('/content/car_price_prediction_with_missing.csv')
print(dataset)
```

### Checking Missing Values

```python
print(dataset.isnull().sum())
print(dataset['Brand'].isnull().sum())
```

### Missing Value Removal

```python
print(dataset.dropna(subset=['Brand'], inplace=True))
print(len(dataset))
print(dataset['Brand'].isnull().sum())
```

### Data Standardization

```python
dataset['Brand']=dataset['Brand'].str.strip()
print(dataset['Brand'])
dataset['Brand']=dataset['Brand'].str.title()
print(dataset['Brand'])
dataset['Brand'] = dataset['Brand'].replace("Bmw", "BMW")
print(dataset['Brand'])
print(dataset['Brand'].unique())
```

### Duplicate Record Detection

```python
print(dataset.duplicated())
print(dataset.duplicated().sum())
```

### Data Integrity Validation

```python
print(dataset['Brand'].isnull().sum())
print(dataset.duplicated().sum())
print(len(dataset['Brand']))
```

### Data Exploration

```python
print(dataset['Brand'])
print(dataset['Brand'].value_counts)
```

### Cleaned Data Extraction

```python
from google.colab import files
dataset.to_csv("Brand_col_cleaned.csv", index=False)
files.download("Brand_col_cleaned.csv")
```
