### Loading the Cleaned Dataset

```python
import pandas as pd
dataset=pd.read_csv('/content/mileage_cleaned.csv')
print(dataset)
```

### Checking Values

```python
print(dataset["Condition"].isnull().sum())
print(dataset["Condition"].dtype)
print(dataset["Condition"].unique())
print(dataset["Condition"].value_counts())
```

### Cleaning and Standardizing Data

```python
dataset["Condition"] = dataset["Condition"].str.strip()
dataset["Condition"] = dataset["Condition"].str.title()
print(dataset["Condition"].unique())
print(dataset["Condition"].value_counts())
```

### Reviewing the Updated Condition Column

```python
print(dataset["Condition"].isnull().sum())
print(dataset["Condition"].dtype)
print(dataset["Condition"].unique())
print(dataset["Condition"].value_counts())
print("Total rows:", len(dataset))
```

### Saving the Final Dataset

```python
from google.colab import files
dataset.to_csv('condition_cleaned.csv',index=False)
files.download('/content/condition_cleaned.csv')
```
