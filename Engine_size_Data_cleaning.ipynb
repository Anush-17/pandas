## Import Dataset

```python
import pandas as pd
dataset=pd.read_csv('/content/Year_col_cleaned.csv')
print(dataset)
```

## Check Missing Values

```python
print(dataset['Engine Size'].isnull().sum())
print(dataset['Engine Size'].dtype)
```

## Display Unique Values

```python
print(dataset['Engine Size'].describe())
print(sorted(dataset['Engine Size'].unique()))
```

## Range and Check Dataset Duplicates

```python
print(dataset["Engine Size"].min())
print(dataset["Engine Size"].max())
print(dataset.duplicated().sum())
print(len(dataset))
```
