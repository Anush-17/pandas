### Loading the Cleaned Dataset

```python
import pandas as pd
import numpy as np
dataset=pd.read_csv('/content/Transmission_cleaned.csv')
print(dataset)
```

### Checking Statistics

```python
print(dataset["Mileage"].isnull().sum())
print(dataset["Mileage"].dtype)
print(dataset["Mileage"].describe())
```

### Finding Outliers in Mileage Data

```python

Q1 = dataset["Mileage"].quantile(0.25)
Q3 = dataset["Mileage"].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower boundary:", lower)
print("Upper boundary:", upper)
outliers = dataset[(dataset["Mileage"] < lower) | (dataset["Mileage"] > upper)]
print("Number of outliers:", len(outliers))
```

### Reviewing the Dataset After Validation

```python
print(dataset["Mileage"].isnull().sum())
print(dataset["Mileage"].dtype)
print(dataset["Mileage"].min())
print(dataset["Mileage"].max())
print(dataset.duplicated().sum())
print(len(dataset))
```
