### Import Dataset

```python
import pandas as pd
dataset=pd.read_csv('/content/Engine_Size_cleaned.csv')
print(dataset)
```

### Checking Fuel Type Column Details

```python
print(dataset["Fuel Type"].isnull().sum())
print(dataset["Fuel Type"].dtype)
print(dataset["Fuel Type"].unique())
print(dataset["Fuel Type"].value_counts())
```

### Cleaning the Fuel Type Values

```python
dataset['Fuel Type']=dataset['Fuel Type'].str.strip()
dataset['Fuel Type']=dataset['Fuel Type'].str.title()
print(dataset['Fuel Type'])
```

### Checking the Data After Cleaning

```python
print(dataset["Fuel Type"].isnull().sum())
print(dataset["Fuel Type"].dtype)
print(dataset["Fuel Type"].unique())
print(dataset.duplicated().sum())
print(len(dataset))
```
